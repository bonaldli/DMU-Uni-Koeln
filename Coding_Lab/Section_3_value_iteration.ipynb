## Import the necessary software libraries:

```python
import numpy as np
import matplotlib.pyplot as plt

from envs import Maze
from utils import plot_policy, plot_values, test_agent
```

## Initialize the environment

```python
env = Maze()
```

```python
frame = env.render(mode='rgb_array')
plt.axis('off')
plt.imshow(frame)
```

```python
print(f'Observation space shape:{env.observation_space.nvec}')
print(f'Number of actions:{env.action_space.n}')
```

## Define the policy $\pi(\cdot|s)$

#### Create the policy $\pi(\cdot|s)$

```python
policy_probs = np.full((5,5,4),0.25) # 25 states with 4 possible actions [0.25, 0.25, 0.25, 0.25]
```

```python
def policy(state):
    return policy_probs[state]
```

#### Test the policy with state (0, 0)

```python
action_probabilities = policy((0,0))
for action, prob in zip(range(4), action_probabilities):
    print(f"Probability of taking action {action}: {prob}")
```

#### See how the random policy does in the maze

```python
test_agent(env, policy, episodes=1)
```

#### Plot the policy

```python
plot_policy(policy_probs, frame)
```

## Define value table $V(s)$

#### Create the $V(s)$ table

```python
state_values = np.zeros(shape=(5,5))
```

#### Plot $V(s)$

```python
plot_values(state_values, frame)
```

## Implement the Value Iteration algorithm

</br>




<div style="text-align:center">
    Adapted from Barto & Sutton: "Reinforcement Learning: An Introduction".
</div>

```python
def value_iteration(policy_probs, state_values, theta=1e-20, gamma=0.99):
    delta = float("inf")
    i = 0
    while delta > theta:
        delta = 0
        for row in range(5):
            for col in range(5):
                old_value = state_values[(row, col)]
                action_probs = None
                max_qsa = float("-inf")
                
                for action in range(4):
                    next_state, reward, _, _ = env.simulate_step((row, col), action) #model-based
                    qsa = reward + gamma * state_values[next_state]
                    
                    if qsa > max_qsa:
                        max_qsa = qsa
                        action_probs = np.zeros(4)
                        action_probs[action] = 1.
                        
                state_values[(row, col)] = max_qsa
                #print("state passed")
                policy_probs[(row, col)] = action_probs
                
                delta = max(delta, abs(max_qsa - old_value)) # where the mistake was: not 'deta', but 'delta'
        print(f'Iteration: {i}, with delta = {delta}')
        plot_values(state_values, frame)
        plot_policy(policy_probs, frame)
        i = i + 1
```

```python
value_iteration(policy_probs, state_values)
```
